```python
import matplotlib.pyplot as plt
import NormalCode.MainCode as mc

def initValues():
    tau = 0.0001
    n = 3

    r = [[1., 1., 0.],
         [0., 0., 0.],
         [1., 0., 0.]]

    v = [[-1., 0., 0.],
         [0., 0., 0.],
         [0., 1., 0.]]

    m = [1., 1., 1.]

    return tau, n, m, r, v

def plotTrajectories(tau, n, m, r, v, num_steps=650000):
    
    # Lists to store trajectories
    x1, y1 = [r[0][0]], [r[0][1]]
    x2, y2 = [r[1][0]], [r[1][1]]
    x3, y3 = [r[2][0]], [r[2][1]]
    
    # Update positions and velocities iteratively
    for _ in range(num_steps):
        r, v = mc.do_step(tau, n, m, r, v)
        x1.append(r[0][0])
        y1.append(r[0][1])
        x2.append(r[1][0])
        y2.append(r[1][1])
        x3.append(r[2][0])
        y3.append(r[2][1])
    
    # Plotting the trajectories
    plt.figure(figsize=(4, 3))
    plt.plot(x1, y1, label='Body 1')
    plt.plot(x2, y2, label='Body 2')
    plt.plot(x3, y3, label='Body 3')
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    plt.title('Trajectories of three bodies')
    plt.legend()
    plt.grid(True)
    plt.axis('equal')
    plt.show()

```

```python
tau, n, m, r, v = initValues()
plotTrajectories(tau, n, m, r, v, num_steps=50000)
```
